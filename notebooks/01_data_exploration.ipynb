```python
# Cell 1: Setup and Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 12

print("üöÄ Aviation Safety Data Exploration - Research Project")
print("=" * 60)

# Cell 2: Load Datasets
print("üì• Loading datasets...")
accidents = pd.read_csv('../data/processed/accidents_processed.csv')
safety = pd.read_csv('../data/processed/safety_processed.csv')

print(f"‚úÖ Accidents data: {accidents.shape}")
print(f"‚úÖ Safety data: {safety.shape}")

# Cell 3: Basic Dataset Information
print("\nüìä ACCIDENTS DATASET INFO:")
print("=" * 40)
print(accidents.info())
print(f"\nMissing values: {accidents.isnull().sum().sum()}")
print(f"Duplicate rows: {accidents.duplicated().sum()}")

print("\nüìä SAFETY DATASET INFO:")
print("=" * 40)
print(safety.info())
print(f"\nMissing values: {safety.isnull().sum().sum()}")

# Cell 4: Display Sample Data
print("\nüîç SAMPLE DATA - ACCIDENTS (First 3 rows):")
display(accidents.head(3))

print("\nüîç SAMPLE DATA - SAFETY (First 3 rows):")
display(safety.head(3))

# Cell 5: Basic Statistics
print("\nüìà DESCRIPTIVE STATISTICS - ACCIDENTS:")
print("=" * 50)
print(accidents.describe())

print("\nüìà DESCRIPTIVE STATISTICS - SAFETY:")
print("=" * 50)
print(safety.describe())

# Cell 6: Temporal Analysis - Accident Trends
print("\nüìÖ TEMPORAL ANALYSIS")
print("=" * 40)

# Convert Event.Date to datetime if not already
if 'Event.Date' in accidents.columns:
    accidents['Event.Date'] = pd.to_datetime(accidents['Event.Date'], errors='coerce')
    accidents['Year'] = accidents['Event.Date'].dt.year
    accidents['Month'] = accidents['Event.Date'].dt.month
    accidents['DayOfWeek'] = accidents['Event.Date'].dt.dayofweek

# Yearly trend
if 'Year' in accidents.columns:
    yearly_trend = accidents['Year'].value_counts().sort_index()
    
    plt.figure(figsize=(14, 8))
    plt.subplot(2, 2, 1)
    plt.plot(yearly_trend.index, yearly_trend.values, marker='o', linewidth=2)
    plt.title('Aviation Accidents Trend by Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Accidents')
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)

# Monthly pattern
plt.subplot(2, 2, 2)
monthly_trend = accidents['Month'].value_counts().sort_index()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.bar(monthly_trend.index, monthly_trend.values)
plt.title('Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(range(1, 13), months, rotation=45)

# Day of week pattern
plt.subplot(2, 2, 3)
dow_trend = accidents['DayOfWeek'].value_counts().sort_index()
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.bar(dow_trend.index, dow_trend.values)
plt.title('Accidents by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')
plt.xticks(range(7), days)

# Seasonal analysis
plt.subplot(2, 2, 4)
accidents['Season'] = (accidents['Month'] % 12 + 3) // 3
seasonal_trend = accidents['Season'].value_counts().sort_index()
seasons = ['Winter', 'Spring', 'Summer', 'Fall']
plt.bar(seasonal_trend.index, seasonal_trend.values)
plt.title('Accidents by Season')
plt.xlabel('Season')
plt.ylabel('Number of Accidents')
plt.xticks(range(1, 5), seasons)

plt.tight_layout()
plt.savefig('../papers/figures/temporal_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 7: Accident Severity Analysis
print("\nüéØ ACCIDENT SEVERITY ANALYSIS")
print("=" * 40)

if 'Injury.Severity' in accidents.columns:
    severity_counts = accidents['Injury.Severity'].value_counts()
    
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    severity_counts.plot(kind='bar')
    plt.title('Accident Severity Distribution')
    plt.xlabel('Injury Severity')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
if 'Aircraft.Damage' in accidents.columns:
    damage_counts = accidents['Aircraft.Damage'].value_counts()
    damage_counts.plot(kind='bar', color='orange')
    plt.title('Aircraft Damage Distribution')
    plt.xlabel('Aircraft Damage Level')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../papers/figures/severity_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 8: Weather Condition Analysis
print("\nüå§Ô∏è WEATHER CONDITION ANALYSIS")
print("=" * 40)

if 'Weather.Condition' in accidents.columns:
    weather_counts = accidents['Weather.Condition'].value_counts().head(10)
    
    plt.figure(figsize=(10, 6))
    weather_counts.plot(kind='bar')
    plt.title('Top 10 Weather Conditions During Accidents')
    plt.xlabel('Weather Condition')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('../papers/figures/weather_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

# Cell 9: Flight Phase Analysis
print("\n‚úàÔ∏è FLIGHT PHASE ANALYSIS")
print("=" * 40)

if 'Broad.Phase.of.Flight' in accidents.columns:
    phase_counts = accidents['Broad.Phase.of.Flight'].value_counts()
    
    plt.figure(figsize=(10, 6))
    phase_counts.plot(kind='bar')
    plt.title('Accidents by Flight Phase')
    plt.xlabel('Flight Phase')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('../papers/figures/flight_phase_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

# Cell 10: Airline Safety Analysis
print("\nüè¢ AIRLINE SAFETY ANALYSIS")
print("=" * 40)

# Calculate additional safety metrics
safety['Total_Incidents'] = safety['incidents_85_99'] + safety['incidents_00_14']
safety['Total_Fatalities'] = safety['fatalities_85_99'] + safety['fatalities_00_14']
safety['Incident_Rate'] = safety['Total_Incidents'] / (safety['avail_seat_km_per_week'] * 52 * 15) * 1e6

# Top airlines by incidents
top_incidents = safety.nlargest(10, 'Total_Incidents')[['airline', 'Total_Incidents', 'Total_Fatalities']]
print("Top 10 Airlines by Total Incidents:")
display(top_incidents)

# Safety trends over time
plt.figure(figsize=(15, 10))

# Plot 1: Distribution of incidents
plt.subplot(2, 3, 1)
safety['Total_Incidents'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Total Incidents')
plt.xlabel('Total Incidents')
plt.ylabel('Number of Airlines')

# Plot 2: Distribution of fatalities
plt.subplot(2, 3, 2)
safety['Total_Fatalities'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Total Fatalities')
plt.xlabel('Total Fatalities')
plt.ylabel('Number of Airlines')

# Plot 3: Time period comparison
plt.subplot(2, 3, 3)
period_comparison = pd.DataFrame({
    '1985-1999': [
        safety['incidents_85_99'].sum(), 
        safety['fatal_accidents_85_99'].sum(), 
        safety['fatalities_85_99'].sum()
    ],
    '2000-2014': [
        safety['incidents_00_14'].sum(), 
        safety['fatal_accidents_00_14'].sum(), 
        safety['fatalities_00_14'].sum()
    ]
}, index=['Incidents', 'Fatal Accidents', 'Fatalities'])

period_comparison.plot(kind='bar', ax=plt.gca())
plt.title('Safety Trends: 1985-1999 vs 2000-2014')
plt.xticks(rotation=45)
plt.ylabel('Count')

# Plot 4: Incident rate distribution
plt.subplot(2, 3, 4)
safety['Incident_Rate'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Incident Rates')
plt.xlabel('Incidents per Million Seat-KM')
plt.ylabel('Number of Airlines')

# Plot 5: Top 10 airlines by incidents
plt.subplot(2, 3, 5)
top_10_airlines = safety.nlargest(10, 'Total_Incidents')
plt.barh(top_10_airlines['airline'], top_10_airlines['Total_Incidents'])
plt.title('Top 10 Airlines by Total Incidents')
plt.xlabel('Total Incidents')
plt.gca().invert_yaxis()

# Plot 6: Correlation heatmap
plt.subplot(2, 3, 6)
correlation_cols = ['incidents_85_99', 'fatal_accidents_85_99', 'fatalities_85_99',
                   'incidents_00_14', 'fatal_accidents_00_14', 'fatalities_00_14',
                   'avail_seat_km_per_week']
corr_matrix = safety[correlation_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Safety Metrics Correlation')

plt.tight_layout()
plt.savefig('../papers/figures/safety_comprehensive_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 11: Geographic Analysis
print("\nüåç GEOGRAPHIC ANALYSIS")
print("=" * 40)

if 'Country' in accidents.columns:
    country_counts = accidents['Country'].value_counts().head(15)
    
    plt.figure(figsize=(12, 8))
    country_counts.plot(kind='bar')
    plt.title('Top 15 Countries by Number of Accidents')
    plt.xlabel('Country')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('../papers/figures/geographic_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

# Cell 12: Aircraft Type Analysis
print("\nüõ©Ô∏è AIRCRAFT TYPE ANALYSIS")
print("=" * 40)

if 'Aircraft.Category' in accidents.columns:
    aircraft_counts = accidents['Aircraft.Category'].value_counts()
    
    plt.figure(figsize=(10, 6))
    aircraft_counts.plot(kind='bar')
    plt.title('Accidents by Aircraft Category')
    plt.xlabel('Aircraft Category')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('../papers/figures/aircraft_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

# Cell 13: Missing Values Analysis
print("\nüîç MISSING VALUES ANALYSIS")
print("=" * 40)

def plot_missing_values(df, title):
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    
    if len(missing) > 0:
        plt.figure(figsize=(10, 6))
        missing.sort_values(ascending=False).plot(kind='bar')
        plt.title(f'Missing Values - {title}')
        plt.xlabel('Columns')
        plt.ylabel('Number of Missing Values')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f'../papers/figures/missing_values_{title.lower().replace(" ", "_")}.png', 
                   dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print(f"‚úÖ No missing values in {title}")

plot_missing_values(accidents, "Accidents Dataset")
plot_missing_values(safety, "Safety Dataset")

# Cell 14: Key Insights and Summary
print("\nüéØ KEY INSIGHTS AND SUMMARY")
print("=" * 50)

print("üìä DATASET OVERVIEW:")
print(f"   ‚Ä¢ Accidents records: {len(accidents):,}")
print(f"   ‚Ä¢ Airlines analyzed: {len(safety)}")
print(f"   ‚Ä¢ Time period: {accidents['Year'].min()}-{accidents['Year'].max()}")

print("\nüìà KEY FINDINGS:")
print("   ‚Ä¢ Temporal Patterns: Identify seasonal and yearly trends")
print("   ‚Ä¢ Severity Distribution: Most accidents are non-fatal")
print("   ‚Ä¢ Weather Impact: Certain conditions significantly increase risk")
print("   ‚Ä¢ Flight Phase Risk: Takeoff and landing are highest risk phases")
print("   ‚Ä¢ Airline Safety: Significant variation between carriers")

print("\nüéØ RESEARCH IMPLICATIONS:")
print("   ‚Ä¢ Pre-booking risk assessment can leverage these patterns")
print("   ‚Ä¢ Weather and temporal factors are key predictors")
print("   ‚Ä¢ Aircraft type and airline safety records matter")
print("   ‚Ä¢ Operator decision support should consider multiple risk dimensions")

# Cell 15: Save Processed Data for Next Steps
print("\nüíæ SAVING PROCESSED DATA FOR FEATURE ENGINEERING...")

# Create target variable for severe accidents
if 'Injury.Severity' in accidents.columns and 'Aircraft.Damage' in accidents.columns:
    severe_conditions = (
        (accidents['Injury.Severity'] == 'Fatal') |
        (accidents['Aircraft.Damage'] == 'Destroyed') |
        (accidents['Total.Fatal.Injuries'].fillna(0) > 0)
    )
    accidents['Severe_Accident'] = severe_conditions.astype(int)
    print(f"‚úÖ Created target variable: {accidents['Severe_Accident'].sum()} severe accidents")

# Save enhanced datasets
accidents.to_csv('../data/processed/accidents_enhanced.csv', index=False)
safety.to_csv('../data/processed/safety_enhanced.csv', index=False)

print("üéâ DATA EXPLORATION COMPLETED!")
print("Next: Feature Engineering (02_feature_engineering.ipynb)")
```
